```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from datetime import datetime, timezone
```

```python
# Save data in a DataFrame
data = pd.read_csv('/content/data.csv')
```

```python
# Delete row with no duration
data = data[data["duration"] != "P0D"].reset_index(drop=True)
```

```python
# Clean date column
data['publishedAt'] = pd.to_datetime(data['publishedAt'])
```

```python
# Function to clean duration column
import re

def iso8601_to_seconds(duration):
    match = re.match(r'PT(?:(\d+)H)?(?:(\d+)M)?(?:(\d+)S)?', duration)
    if not match:
        return 0
    h, m, s = match.groups()
    return int(h or 0) * 3600 + int(m or 0) * 60 + int(s or 0)
```

```python
# Add duration columns
data["duration_seconds"] = data["duration"].apply(iso8601_to_seconds)
data["duration_minutes"] = data["duration_seconds"] / 60
data["duration_hours"]   = data["duration_seconds"] / 3600
```
